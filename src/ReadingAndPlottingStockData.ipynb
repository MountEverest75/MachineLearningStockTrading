```python
import os
def get_max_close(ticker_symbol):
    """Return the Maximum Closing Price of a stock"""
    #Dynamically read a ticker symbol's csv file
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(ticker_symbol))))
    return df['Close'].max()

def get_mean_volume(symbol):
    """Return the mean volume for stock indicated by symbol.
    
    Note: Data for a stock is stored in file: data/<symbol>.csv
    """
    base_dir = "../data/"
    #print(os.path.join(base_dir, "{}.csv".format(str(ticker))))
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(symbol))))  # read in data
    # TODO: Compute and return the mean volume for this stock
    return df['Volume'].mean()


for symbol in ['AMZN', 'JNJ']:
    print("Max Close:",symbol,get_max_close(symbol))
    print("Mean Volume:",symbol,get_mean_volume(symbol))
```

```python
# Bug fixes to join data frames. Dynamically fetch the start date of trading before joining and indexing closing prices by date
from datetime import datetime, date

def get_stock_datadir(ticker,base_dir="../data/"):
    #Return CSV file given ticker symbol
    return os.path.join(base_dir, "{}.csv".format(str(ticker)))

def get_date_range_df(ticker):
    normal_df = pd.read_csv(get_stock_datadir(ticker))
    start_date = normal_df['Date'].iloc[0]
    #end_date = date(date.today().year, date.today().month, date.today().day) # Gets current record
    end_date = normal_df['Date'].iloc[-1] #Gets the end date from last record
    dates = pd.date_range(start_date, end_date)
    df_dates = pd.DataFrame(index=dates)
    return df_dates
    

def join_data_frames(ticker):
    #Create Dataframe for ticker symbol using Date column as index. Also indicate that dates need to be parsed
    base_dir="../data/"
    df_ticker = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(ticker))),index_col="Date", parse_dates=True, usecols=['Date', 'Close', 'Adj Close'],na_values=['nan'])
    df_dates = get_date_range_df(ticker)
    df = df_dates.join(df_ticker, how='inner')
    return df
    
    
print("For JNJ:")
print(join_data_frames('JNJ'))
print("For AMZN:")
print(join_data_frames('AMZN'))
```

```python
def join_multiple_tickers(tickers):
    #Create empty data frames with date range
    base_dir = "../data/"
    df_old = pd.DataFrame()
    for symbol in tickers:
        df = get_date_range_df(symbol)
        df_temp = pd.read_csv(os.path.join(base_dir, "{}.csv".format(symbol)), index_col="Date", parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp, how='inner')
        if not df_old.empty:
            df = df_old.join(df,how='inner')
        df_old = df
    df = df.dropna()
    return df

tickers = ['AMZN', 'JNJ', 'SPY']
#tickers = ['AMZN']
join_multiple_tickers(tickers)
```

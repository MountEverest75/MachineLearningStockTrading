```python
# Bug fixes to join data frames. Dynamically fetch the start date of trading before joining and indexing closing prices by date
from datetime import datetime, date

def get_stock_datadir(ticker,base_dir="../data/"):
    #Return CSV file given ticker symbol
    return os.path.join(base_dir, "{}.csv".format(str(ticker)))

def get_date_range_df(ticker):
    normal_df = pd.read_csv(get_stock_datadir(ticker))
    start_date = normal_df['Date'].iloc[0]
    #end_date = date(date.today().year, date.today().month, date.today().day) # Gets current record
    end_date = normal_df['Date'].iloc[-1] #Gets the end date from last record
    dates = pd.date_range(start_date, end_date)
    df_dates = pd.DataFrame(index=dates)
    return df_dates
    

def join_data_frames(ticker):
    #Create Dataframe for ticker symbol using Date column as index. Also indicate that dates need to be parsed
    base_dir="../data/"
    df_ticker = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(ticker))),index_col="Date", parse_dates=True, usecols=['Date', 'Close', 'Adj Close'],na_values=['nan'])
    df_dates = get_date_range_df(ticker)
    df = df_dates.join(df_ticker, how='inner')
    return df
    
    
print("For JNJ:")
print(join_data_frames('JNJ'))
print("For AMZN:")
print(join_data_frames('AMZN'))
```

```python
def join_multiple_tickers(tickers):
    #Create empty data frames with date range
    base_dir = "../data/"
    df_old = pd.DataFrame()
    for symbol in tickers:
        df = get_date_range_df(symbol)
        df_temp = pd.read_csv(os.path.join(base_dir, "{}.csv".format(symbol)), index_col="Date", parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp, how='inner')
        if not df_old.empty:
            df = df_old.join(df,how='inner')
        df_old = df
    df = df.dropna()
    return df

tickers = ['AMZN', 'JNJ', 'SPY']
#tickers = ['AMZN']
join_multiple_tickers(tickers)
```

## Rolling Statistics and Bollinger Bands

```python
# Rolling Statistics - Collected on a time series window. Duration of days etc. 200 day moving average is an example
# Rolling mean also called moving average is common technical indicator. Rolling Standard Deviation and Bollinger bands show how significant is the deviation observed.
def symbol_to_path(symbol, base_dir="../data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))


def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SPY')

    for symbol in symbols:
        df_temp = pd.read_csv(symbol_to_path(symbol), index_col='Date',
                parse_dates=True, usecols=['Date', 'Adj Close'], na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp)
        if symbol == 'SPY':  # drop dates SPY did not trade
            df = df.dropna(subset=["SPY"])

    return df


def plot_data(df, title="Stock prices"):
    """Plot stock prices with a custom title and meaningful axis labels."""
    ax = df.plot(title=title, fontsize=12)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()


def get_rolling_mean(values, window):
    """Return rolling mean of given values, using specified window size."""
    series = pd.Series(values)
    return series.rolling(window).mean()


def get_rolling_std(values, window):
    """Return rolling standard deviation of given values, using specified window size."""
    # TODO: Compute and return rolling standard deviation
    series = pd.Series(values)
    return series.rolling(window).std()
    


def get_bollinger_bands(rm, rstd):
    """Return upper and lower Bollinger Bands."""
    # TODO: Compute upper_band and lower_band
    lower_band = rm - rstd*2
    upper_band = rm + rstd*2
    return upper_band, lower_band


def test_run():
    # Read data
    dates = pd.date_range('1950-01-01', end_date)
    symbols = ['AMZN','JNJ','SPY']
    df = get_data(symbols, dates)

    # Compute Bollinger Bands
    # 1. Compute rolling mean
    #rm_SPY = get_rolling_mean(df['SPY'], window=20)
    rm_SPY = get_rolling_mean(df['AMZN'], window=20)

    # 2. Compute rolling standard deviation
    #rstd_SPY = get_rolling_std(df['SPY'], window=20)
    rstd_SPY = get_rolling_std(df['AMZN'], window=20)

    # 3. Compute upper and lower bands
    upper_band, lower_band = get_bollinger_bands(rm_SPY, rstd_SPY)
    
    # Plot raw SPY values, rolling mean and Bollinger Bands
    ax = df['AMZN'].plot(title="Bollinger Bands", label='AMZN')
    rm_SPY.plot(label='Rolling mean', ax=ax)
    upper_band.plot(label='upper band', ax=ax)
    lower_band.plot(label='lower band', ax=ax)

    # Add axis labels and legend
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    ax.legend(loc='upper left')
    plt.show()
    
    # Compute Bollinger Bands
    # 1. Compute rolling mean
    rm_SPY = get_rolling_mean(df['SPY'], window=20)
    
    # 2. Compute rolling standard deviation
    rstd_SPY = get_rolling_std(df['SPY'], window=20)

    # 3. Compute upper and lower bands
    upper_band, lower_band = get_bollinger_bands(rm_SPY, rstd_SPY)
    
    # Plot raw SPY values, rolling mean and Bollinger Bands
    ax = df['SPY'].plot(title="Bollinger Bands", label='SPY')
    rm_SPY.plot(label='Rolling mean', ax=ax)
    upper_band.plot(label='upper band', ax=ax)
    lower_band.plot(label='lower band', ax=ax)

    # Add axis labels and legend
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    ax.legend(loc='upper left')
    plt.show()
    
    # Compute Bollinger Bands
    # 1. Compute rolling mean
    rm_SPY = get_rolling_mean(df['JNJ'], window=20)
    
    # 2. Compute rolling standard deviation
    rstd_SPY = get_rolling_std(df['JNJ'], window=20)

    # 3. Compute upper and lower bands
    upper_band, lower_band = get_bollinger_bands(rm_SPY, rstd_SPY)
    
    # Plot raw SPY values, rolling mean and Bollinger Bands
    ax = df['JNJ'].plot(title="Bollinger Bands", label='JNJ')
    rm_SPY.plot(label='Rolling mean', ax=ax)
    upper_band.plot(label='upper band', ax=ax)
    lower_band.plot(label='lower band', ax=ax)

    # Add axis labels and legend
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    ax.legend(loc='upper left')
    plt.show()


    

if __name__ == "__main__":
    test_run()
```

```python
# daily_returns = df.copy()
# daily_returns[1:] = (df[1:] / df[:-1].values) - 1
# daily_returns.iloc[0, :] = 0

def plot_data(df, title="Stock prices", xlabel="Date", ylabel="Price"):
    """Plot stock prices with a custom title and meaningful axis labels."""
    ax = df.plot(title=title, fontsize=12)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()

# Using Pandas Data Frame
def compute_daily_returns(df):
    """Compute and return the daily return values."""
    # TODO: Your code here
    # Note: Returned DataFrame must have the same number of rows
    daily_returns = (df / df.shift(1)) - 1
    daily_returns.iloc[0, :] = 0
    return daily_returns

def test_run():
    # Read data
    dates = pd.date_range('1950-01-01', end_date)  # one month only
    symbols = ['SPY','AMZN']
    df = get_data(symbols, dates)
    plot_data(df)

    # Compute daily returns
    daily_returns = compute_daily_returns(df)
    plot_data(daily_returns, title="Daily returns", ylabel="Daily returns")


if __name__ == "__main__":
    test_run()
```

```python
"""Fill missing values"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

def fill_missing_values(df_data):
    """Fill missing values in data frame, in place."""
    ##########################################################
    #pass  # TODO: Your code here (DO NOT modify anything else)
    ##########################################################
    # df_data.fillna(method='ffill')
    df_data.fillna(method="ffill", inplace=True) 
    df_data.fillna(method="bfill", inplace=True) 


def symbol_to_path(symbol, base_dir="../data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))


def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df_final = pd.DataFrame(index=dates)
    if "SPY" not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, "SPY")

    for symbol in symbols:
        file_path = symbol_to_path(symbol)
        df_temp = pd.read_csv(file_path, parse_dates=True, index_col="Date",
            usecols=["Date", "Adj Close"], na_values=["nan"])
        df_temp = df_temp.rename(columns={"Adj Close": symbol})
        df_final = df_final.join(df_temp)
        if symbol == "SPY":  # drop dates SPY did not trade
            df_final = df_final.dropna(subset=["SPY"])

    return df_final


def plot_data(df_data):
    """Plot stock data with appropriate axis labels."""
    ax = df_data.plot(title="Stock Data", fontsize=2)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()


def test_run():
    """Function called by Test Run."""
    # Read data
    symbol_list = ["AMZN", "JNJ", "SPY"]  # list of symbols
    start_date = "1950-01-01"
    #end_date = "2014-12-07"
    dates = pd.date_range(start_date, end_date)  # date range as index
    df_data = get_data(symbol_list, dates)  # get data for each symbol

    # Fill missing values
    fill_missing_values(df_data)

    # Plot
    plot_data(df_data)


if __name__ == "__main__":
    test_run()
```

```python
import os
def get_max_close(ticker_symbol):
    """Return the Maximum Closing Price of a stock"""
    #Dynamically read a ticker symbol's csv file
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(ticker_symbol))))
    return df['Close'].max()

def get_mean_volume(symbol):
    """Return the mean volume for stock indicated by symbol.
    
    Note: Data for a stock is stored in file: data/<symbol>.csv
    """
    base_dir = "../data/"
    #print(os.path.join(base_dir, "{}.csv".format(str(ticker))))
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(str(symbol))))  # read in data
    # TODO: Compute and return the mean volume for this stock
    return df['Volume'].mean()


for symbol in ['AMZN', 'JNJ']:
    print("Max Close:",symbol,get_max_close(symbol))
    print("Mean Close:",symbol,get_mean_volume(symbol))
```

```python
def plot_price_history(ticker):
    """Plot price history in reverse order"""
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(ticker)))
    df['Adj Close'].plot()
    plt.show() #call to show plots
    
for symbol in ['AMZN', 'JNJ']:
    plot_price_history(symbol)
```

```python
def plot_high_history(ticker):
    """Plot price history in reverse order"""
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(ticker)))
    print(df['High'])
    df['High'].plot()
    plt.show() #call to show plots
    
for symbol in ['AMZN', 'JNJ']:
    plot_high_history(symbol)
```

```python
def plot_two_columns(ticker):
    """Plot two portions of data"""
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(ticker)))
    df[['Close', 'Adj Close']].plot()
    plt.show()

for symbol in ['AMZN', 'JNJ']:
    plot_two_columns(symbol)
```

```python
def get_moving_average(ticker, duration):
    """Example to get 200 day moving average, pass duration=200 and ticker symbol=JNJ or any stock"""
    base_dir = "../data/"
    df = pd.read_csv(os.path.join(base_dir, "{}.csv".format(ticker)))
    df = df[1:201]
    return df['Close'].mean()

for symbol in ['AMZN', 'JNJ']:
    duration=200
    print("Moving Average for: ", duration, "days. For ticker:", symbol, get_moving_average(symbol,duration))
```

```python
from datetime import datetime, date
def get_stock_datadir(ticker,base_dir="../data/"):
    """Return CSV file given ticker symbol"""
    return os.path.join(base_dir, "{}.csv".format(str(ticker)))

def get_date_range_df():
    start_date = '2010-01-01' #TBD - Receive Ticker symbol and setup start date as date the first time ticker traded
    end_date = date(date.today().year, date.today().month, date.today().day)
    dates = pd.date_range(start_date, end_date)
    df_dates = pd.DataFrame(index=dates)
    return df_dates

#df_dates = get_date_range_df()

def join_data_frames(ticker):
    #Create empty data frames with date range
    df_dates = get_date_range_df()
    
    #Create Dataframe for ticker symbol using Date column as index. Also indicate that dates need to be parsed
    base_dir="../data/"
    df_ticker = pd.read_csv(os.path.join(base_dir, "{}.csv".format(ticker)), index_col="Date", parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
    #df = df_dates.join(df_ticker)
    #df = df.dropna()
    #The how parameter allows to join in one statement
    df = df_dates.join(df_ticker, how='inner')
    return df
    
    
print(join_data_frames('SPY'))
```

```python
def join_multiple_tickers(tickers):
    #Create empty data frames with date range
    df_dates = get_date_range_df()
    df = df_dates
    base_dir = "../data/"
    for symbol in tickers:
        df_temp = pd.read_csv(os.path.join(base_dir, "{}.csv".format(symbol)), index_col="Date", parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df_ticker = df_dates.join(df_temp, how='inner')
        df = df.join(df_ticker)
    df = df.dropna()
    return df

tickers = ['AMZN', 'JNJ', 'SPY']
#tickers = ['AMZN']
join_multiple_tickers(tickers)
```

```python
def symbol_to_path(symbol, base_dir="../data/"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))

def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SPY')

    base_dir = "../data/"
    for symbol in symbols:
        # TODO: Read and join data for each symbol
        df_temp = pd.read_csv(os.path.join(base_dir, "{}.csv".format(symbol)), index_col="Date", parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        #print(df_temp)
        df = df.join(df_temp)
    df = df.dropna()
    #if symbol == 'SPY':
        #df = df.dropna(subset=["SPY"])
        
    return df


def test_run():
    # Define a date range
    dates = pd.date_range('2010-01-01', '2017-10-22')
    #dates = get_date_range_df()

    # Choose stock symbols to read
    symbols = ['AMZN', 'JNJ']
    
    # Get stock data
    df = get_data(symbols, dates)
    print(df)

test_run()
dates = pd.date_range('2010-01-01', '2017-10-22')
symbols = ['AMZN', 'JNJ']
df = get_data(symbols, dates)
```

```python
def plot_data(df, title="Stock Prices"):
    """Plot Data Frame"""
    #Commented statements provide basic plotting
    #df.plot()
    #plt.show()
    #Send Title to the graph. Also the output of df.plot() is a handler "df_plot" to add x and y axis labels
    df_plot = df.plot(title=title)
    df_plot.set_xlabel("Date")
    df_plot.set_ylabel("Price")
    plt.show()
    
plot_data(df)
```

```python
def plot_selected(df, columns, start_index, end_index):
    """Plot the desired columns over index values in the given range."""
    # TODO: Your code here
    # Note: DO NOT modify anything else!
    #Define Range using start:end. Put columns to be selected.
    #df = df.ix[start_index:end_index, columns]
    df = df.loc[start_index:end_index, columns]
    #print(df)
    plot_data(df,title="Selected Tickers")

print(symbols)
df_test = get_data(symbols, dates)
#print(df_test)
plot_selected(df_test, ['AMZN', 'JNJ'], '2017-03-01', '2017-04-01')
```

```python
#How much the stock went up or down when compared to others. This could be done by normalizing the prices to 1.0
def normalize_stock_prices(df):
    """By dividing the value of each stock by price at day one"""
    #return df / df.ix[0, :]
    return df / df.iloc[0, :]
    #return df / df.loc[0, :]

plot_selected(normalize_stock_prices(df_test), ['AMZN', 'JNJ'], '2010-01-01', '2017-10-22')
```

# 0. Set API Keys and Import Libraries

```python
#Import Pandas to Store results into Data Frames
import pandas as pd

keys={}
with open("/Users/mounteverest/Boot_Settings/APIFile.txt") as APIKeys:
    for line in APIKeys:
        name, value = line.split("=")
        keys[name.strip()]=value.strip()

QUANDL_API_KEY = keys['QUANDL_API_KEY']
ALPHA_VANTAGE_API_KEY = keys['ALPHA_VANTAGE_API_KEY']
WORLD_TRADING_API_KEY = keys['WORLD_TRADING_API_KEY']
```

# 1. ALPHA VANTAGE API CALLS

```python
# Use HTTP(S) Endpoints to get data from Alpha Vantage
import requests
#Import OS library to perform common folder and OS operations
import os
#Handle CSV files
import csv

#Make a Alpha Vantage API call using requests object.
alpha_url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=MSFT&apikey="+ALPHA_VANTAGE_API_KEY+"&datatype=csv"
response = requests.get(alpha_url)
file_out = open("../data/MSFT.csv","w")
file_out.write(response.text)
file_out.close()
df_msft = pd.read_csv("../data/MSFT.csv")
print(df_msft)
# print(response.iter_lines())
# for line in response.iter_lines():
#     print(line)
```
